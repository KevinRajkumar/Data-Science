```python
## Importing libraries needed 

import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
# Loading tips dataset which is already present.

tips=sns.load_dataset("tips")
tips.tail()
```

```python
# Descriptive statistics for numeric columns

tips.describe()
```

```python
# To check the number of observations present which has null values or not and datatype of the column

tips.info()
```

```python
# Unique values present in the columns

tips.nunique()
```

# Scatter plot

```python
sns.lmplot(x='total_bill', y='tip',data=tips,scatter=True,height=6,fit_reg=False)
```

#### We can infer that most of the bill lies between 10 dollars to 30 dollars and also tips given is between 1.5 dollars to 4 dollars for the bills. and there are few people whose bill ranges from 40 dollars to 50 dollars.

```python
sns.lmplot(x='total_bill', y='tip',hue='sex',data=tips,scatter=True,fit_reg=False,height=6)
plt.show()
```

#### This is the same type of scatterplot as above but the difference we can find is that most of the males who pay bills between 40 dollars and 50 dollars

```python
sns.lmplot(x='total_bill', y='tip',hue='smoker',col='day',data=tips,scatter=True,fit_reg=False,height=6,palette='prism',markers=['o','x'])
plt.show()
```

#### We can infer from the above scatter plot as people spend most of their money on saturday and sunday because it will be holiday for everyone 
#### and even most of the smokers are found on those days and on friday only a very few people spend money

```python
sns.lmplot(x='total_bill', y='tip',hue='smoker',col='time',row='day',data=tips,scatter=True,fit_reg=False,height=6,palette='prism',markers=['o','x'])
plt.show()
```

#### We can find that only on thursday afternoon, saturday and sunday night people spend money outside.

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False,palette='rainbow',size=4)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,palette='rainbow',size=6)
plt.show()
```

```python
sns.countplot(x='sex',data=tips,palette='Oranges')
plt.show()
```

#### Count of male is higher with approximately 150 where females approximately 80 females spend money.

```python
sns.countplot(x='day',hue='smoker',data=tips,palette='Oranges')
plt.show()
```

#### Smokers do mostly spend on saturday which includes both male and females

```python
sns.distplot(tips['total_bill'])
plt.show()
```

#### From the above graph using bill field We can understand that most of the bill lies between 10 dollars to 30 dollars and there are few outliers with 40 dollars to 60 dollars which makes the data to be skewed towards right side.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,size=4)
plt.show()
```

```python
sns.boxplot(x=tips['total_bill'])
plt.show()
```
